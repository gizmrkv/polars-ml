```python
import lightgbm as lgb
import polars as pl
import polars.selectors as cs
from polars import DataFrame
from sklearn.datasets import load_digits

from polars_ml import Pipeline
from polars_ml.model_selection import KFold, train_test_split
from polars_ml.model_selection.metrics import evaluate_classification_metrics

bunch = load_digits(as_frame=True)
df: DataFrame = pl.from_pandas(bunch.data).with_columns(pl.from_pandas(bunch.target))  # type: ignore
df
```

```python
train_idx, valid_idx = train_test_split(
    df, 0.2, stratify="target", shuffle=True, seed=42
)

train_df = df.select(pl.all().gather(train_idx))
valid_df = df.select(pl.all().gather(valid_idx))

model = (
    Pipeline()
    .with_columns(cs.starts_with("pixel") / 255)
    .concat(
        Pipeline().reduction.umap(
            pl.exclude("target"),
            include_input=False,
            model_kwargs={"n_components": 2},
        ),
        Pipeline().select("target"),
        how="horizontal",
    )
)
model.fit(train_df, valid_df)
train_pred_df = model.transform(train_df)
valid_pred_df = model.transform(valid_df)

train_plotter = (
    Pipeline()
    .with_columns(pl.col("target").cast(pl.String))
    .plot.scatter("umap_0", "umap_1", "target", out_dir="umap_train")
)
train_plotter.transform(train_pred_df)

valid_plotter = (
    Pipeline()
    .with_columns(pl.col("target").cast(pl.String))
    .plot.scatter("umap_0", "umap_1", "target", out_dir="umap_valid")
)
valid_plotter.transform(valid_pred_df)
```

```python
metrics = []
for train_idx, valid_idx in KFold(
    n_splits=3, shuffle=True, seed=42, stratify="target"
).split(df):
    train_df = df.select(pl.all().gather(train_idx))
    valid_df = df.select(pl.all().gather(valid_idx))

    model = (
        Pipeline()
        .with_columns(cs.starts_with("pixel") / 255)
        .concat(
            Pipeline().reduction.umap(
                pl.exclude("target"),
                include_input=False,
                model_kwargs={"n_components": 2},
            ),
            Pipeline().select("target"),
            how="horizontal",
        )
        .gbdt.lightgbm(
            pl.exclude("target"),
            "target",
            {"objective": "multiclass", "num_class": 10},
            train_kwargs={"callbacks": [lgb.early_stopping(10)]},
        )
        .horizontal.argmax(cs.starts_with("lightgbm_"), value_name="lightgbm")
        .with_columns(
            pl.col("lightgbm")
            .list.first()
            .str.extract(r"lightgbm_(\d+)")
            .cast(pl.UInt8)
        )
    )

    model.fit(train_df, valid_df)
    valid_pred_df = model.transform(valid_df)
    metrics.append(
        evaluate_classification_metrics(
            valid_pred_df,
            "target",
            y_pred_class="lightgbm",
            y_pred_proba_prefix="lightgbm_",
            n_classes=10,
        )
    )

pl.concat(metrics)
```
